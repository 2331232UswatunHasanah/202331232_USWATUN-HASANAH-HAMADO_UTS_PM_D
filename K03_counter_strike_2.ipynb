```python
dataset = pd.read_csv('K03_counter_strike_2.csv')
dataset
```

Perintah dataset = pd.read_csv('K03_counter_strike_2.csv') digunakan untuk membaca file CSV dan menyimpannya dalam variabel dataset sebagai DataFrame. Untuk melihat data, kamu bisa menggunakan dataset.head() untuk menampilkan lima baris pertama.

```python
dataset.info()
```

Perintah dataset.info() digunakan untuk menampilkan informasi tentang dataset, seperti:

Jumlah baris dan kolom.
Nama kolom dan tipe data masing-masing.
Jumlah nilai non-null (tidak kosong) pada setiap kolom.
Ini memberikan gambaran umum tentang struktur dataset.

```python
categorical_features = [feature for feature in dataset.columns
                        if dataset[feature].dtypes == "O"]
dataset[categorical_features]
```

Kode tersebut berfungsi untuk:

Membuat daftar categorical_features, yang berisi nama-nama kolom dalam dataset yang memiliki tipe data object (biasanya untuk data kategori atau teks).

dataset[categorical_features] kemudian menampilkan hanya kolom-kolom yang termasuk dalam daftar categorical_features.

Secara sederhana, ini digunakan untuk mengekstrak dan menampilkan kolom yang berisi data kategori (seperti string atau label).

```python
for i in categorical_features:
    print(dataset[i].unique())
```

Kode ini digunakan untuk menampilkan nilai unik yang ada dalam setiap kolom kategori (data tipe object) dalam dataset.

    for i in categorical_features: akan mengiterasi setiap kolom dalam daftar categorical_features.
    dataset[i].unique() akan menampilkan nilai-nilai unik (tanpa duplikasi) dari kolom i.
Dengan kode ini, kamu bisa melihat semua kategori yang terdapat dalam setiap kolom kategori dalam dataset.

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

for feature in categorical_features:
    print(dataset[feature].unique())
    dataset[feature] = label_encoder.fit_transform(dataset[feature])
    print(dataset[feature].unique())
```

Kode ini menggunakan LabelEncoder dari pustaka scikit-learn untuk mengubah nilai kategori (teks) dalam kolom menjadi angka (label numerik):

LabelEncoder: Digunakan untuk mengubah data kategori menjadi angka, yang sering kali diperlukan untuk pemodelan mesin atau algoritma yang memerlukan data numerik.

for feature in categorical_features:: Mengiterasi setiap kolom yang ada dalam daftar categorical_features.

dataset[feature].unique(): Menampilkan nilai unik sebelum transformasi.

label_encoder.fit_transform(dataset[feature]): Mengonversi nilai kategori menjadi angka dan mengganti kolom tersebut dengan nilai yang sudah terkodekan.

dataset[feature].unique(): Menampilkan nilai unik setelah transformasi, yang sekarang berupa angka.

Dengan cara ini, kolom kategori dalam dataset akan diubah menjadi format numerik yang siap digunakan dalam analisis atau pemodelan.

```python
dataset
```

```python
x = dataset.drop(['map', 'round_winner'], axis = 1)
y = dataset['round_winner']
```

Kode ini memisahkan dataset menjadi dua bagian:

x: Fitur, dengan menghapus kolom map dan round_winner.
y: Target, yang berisi kolom round_winner.

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.25,random_state = 0)
```

Kode ini digunakan untuk membagi dataset menjadi data latih (training) dan data uji (test):

train_test_split(x, y, test_size=0.25, random_state=0):
x: Fitur (independent variables).
y: Target (dependent variable).
test_size=0.25: 25% dari data akan digunakan untuk uji, sisanya 75% untuk latih.
random_state=0: Menetapkan nilai acak agar pembagian data konsisten setiap kali kode dijalankan.
Hasilnya:

x_train, y_train: Data latih untuk fitur dan target.
x_test, y_test: Data uji untuk fitur dan target.

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(x_train, y_train)
```

Kode ini digunakan untuk melatih model Naive Bayes menggunakan algoritma Gaussian Naive Bayes:

        from sklearn.naive_bayes import GaussianNB: Mengimpor model Gaussian Naive Bayes dari pustaka scikit-learn.
        
        model = GaussianNB(): Membuat objek model GaussianNB.
        
        model.fit(x_train, y_train): Melatih model menggunakan data latih (x_train untuk fitur dan y_train untuk target).

Model ini akan mempelajari hubungan antara fitur dan target untuk kemudian digunakan dalam prediksi pada data uji.

```python
y_pred_train = model.predict(x_train)
y_pred_test = model.predict(x_test)
```

y_pred_train = model.predict(x_train): Menggunakan model untuk memprediksi target pada data latih (x_train), hasil prediksi disimpan dalam y_pred_train.

y_pred_test = model.predict(x_test): Menggunakan model untuk memprediksi target pada data uji (x_test), hasil prediksi disimpan dalam y_pred_test.

```python
from sklearn.metrics import confusion_matrix, classification_report
print('evaluasi training')
print(confusion_matrix(y_train, y_pred_train))
print(classification_report(y_train, y_pred_train))

print('evaluasi testing')
print(confusion_matrix(y_test, y_pred_test))
print(classification_report(y_test, y_pred_test))
```

Kode ini digunakan untuk mengevaluasi kinerja model menggunakan metrik confusion matrix dan classification report:

confusion_matrix(y_train, y_pred_train): Menghitung dan menampilkan matriks kebingunguan (confusion matrix) untuk data latih. Matriks ini menunjukkan jumlah prediksi yang benar dan salah untuk setiap kelas pada data latih.

classification_report(y_train, y_pred_train): Menampilkan laporan klasifikasi yang mencakup metrik seperti precision, recall, f1-score, dan support untuk data latih.

confusion_matrix(y_test, y_pred_test): Menghitung dan menampilkan matriks kebingunguan untuk data uji.

classification_report(y_test, y_pred_test): Menampilkan laporan klasifikasi untuk data uji.

Dengan kode ini, kamu bisa mengevaluasi seberapa baik model memprediksi pada data latih dan data uji.
